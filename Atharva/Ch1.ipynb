# Finding & removing outliers

```python
df=pd.read_csv('auto-mpg.csv')
df
```

```python
q1=df['mpg'].quantile(0.25)
q1
```

```python
q3=df['mpg'].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
low=q1-(1.5*iqr)
low
```

```python
high=q3+(1.5*iqr)
high
```

```python
finding_outliers=df[(df['acceleration']>high)|(df['acceleration']<low)]
finding_outliers
```

```python
removing_outliers=df[(df['acceleration']<high)&(df['acceleration']>low)]
removing_outliers
```

# Concat

```python
result=pd.concat([df1,df2])
result
```

```python
result=pd.concat([df1,df2],axis=1)
result
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"D":[7,8]})
```

```python
result=pd.concat([df1,df2],join='inner')
result
```

# ignore index

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"D":[7,8]})
result=pd.concat([df1,df2],ignore_index=True)
result
```

# Keys

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})
result=pd.concat([df1,df2],keys=['Train','Test'],ignore_index=True)
result
```

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})
result=pd.concat([df1,df2],keys=['Train','Test'])
result
```

# Names

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})
result=pd.concat([df1,df2],keys=['Train','Test'],names=['DataSet'])
result
```

# verify_integrity

```python
df1=pd.DataFrame({"A":[1,2],"B":[3,4]})
df2=pd.DataFrame({"A":[5,6],"B":[7,8]})

result=pd.concat([df1,df2],verify_integrity=False)
result
```

```python
df1=pd.DataFrame({"E":[1,2],"B":[3,4]})
df2=pd.DataFrame({"C":[5,6],"D":[7,8]})

result=pd.concat([df1,df2],verify_integrity=False,sort=True)
result
```

```python
import pandas as pd
employee=pd.DataFrame({'emp_id':[1,2,3],
                      'name':['A','B','C'],
                      'dept_id':[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],
                         'dept_name':['HR','IT','Finance']})
```

```python
pd.merge(employee,departments,on='dept_id',how='inner')
```

```python
pd.merge(employee,departments,on='dept_id',how='outer')
```

# suffix

```python
import pandas as pd
employee=pd.DataFrame({'emp_id':[1,2,3],
                      'name':['A','B','C'],
                      'dept_id':[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,30],
                         'name':['HR','IT','Finance']})

pd.merge(employee,departments,on='dept_id',suffixes=('_emp','_dept'))
```

```python
import pandas as pd
employee=pd.DataFrame({'emp_id':[1,2,3],
                      'name':['A','B','C'],
                      'dept_id':[10,20,30]})
departments=pd.DataFrame({'dept_id':[10,20,40],
                         'dept_name':['HR','IT','Finance']})
pd.merge(employee,departments,on='dept_id',how='left',indicator=True)
```
