# Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={
    'Math':[80,70,60],
    'Science':[75,65,55],
    'English':[90,85,70]
}
df=pd.DataFrame(data,index=['A','B','C'])
df
```

```python
sns.heatmap(df)
plt.title('Student Marks Heatmap')
plt.show()
```

```python
?sns.heatmap
```

```python
sns.heatmap(df,vmax=50,vmin=100,cmap='YlGnBu')
plt.title('Student Marks Heatmap')
plt.show()
```

```python
sns.heatmap(df,vmax=50,vmin=100,cmap='rocket')
plt.title('Student Marks Heatmap')
plt.show()
```

```python
sns.heatmap(df,vmax=50,vmin=100,cmap=['r','y','g','b','m'])
plt.title('Student Marks Heatmap')
plt.show()
```

```python
sns.heatmap(df,vmax=50,vmin=100,cmap='YlGnBu',linewidths=1,linecolor='k')
plt.title('Student Marks Heatmap')
plt.show()
```

```python
sns.heatmap(df,cmap='YlGnBu',linewidths=1,linecolor='k',annot=True,cbar=False)
plt.title('Student Marks Heatmap')
plt.show()
```

# Pie-chart

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','HR','Finance','Marketing'],
     'Employees':[40,20,25,15]}
df=pd.DataFrame(data)
fig=px.pie(df,values='Employees',names='Department',
          color_discrete_sequence=['blue','green','orange','red'],
          title='Employee Distribution by Department')
fig.update_traces(
    textinfo='percent+label',textposition='inside',pull=[0.111013,0.21,0,0])
fig.show()
```
